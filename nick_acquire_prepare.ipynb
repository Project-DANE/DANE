```python
df = pd.read_csv('cws_residents.csv')
```

```python
pd.options.display.max_columns = None
df.drop(columns=['HMY',"HMYPerson", 'hTenant', 'hTent', 'Rent', 'SRENT'], inplace = True)
```

```python
df
```

```python
#Filter by Current status
df_current = df[df['sStatus'] == 'Current']

# Get the indices of users who only have Past status
idx_past = df[~df['HPerson'].isin(df_current['HPerson'])].index

# Filter rows based on past status for users who only have past status
df_past = df.loc[idx_past]

# Combine the current and past dataframes
df_combined = pd.concat([df_current, df_past])

# Sort by HPerson to get the final result
df_final = df_combined.sort_values('HPerson')
```
